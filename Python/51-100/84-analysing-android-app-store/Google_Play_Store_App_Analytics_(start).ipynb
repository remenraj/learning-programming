# Find the Most Expensive Apps, Filter out the Junk, and Calculate a (ballpark) Sales Revenue Estimate

Let's examine the Price column more closely.

**Challenge**: Convert the price column to numeric data. Then investigate the top 20 most expensive apps in the dataset.

Remove all apps that cost more than $250 from the `df_apps_clean` DataFrame.

Add a column called 'Revenue_Estimate' to the DataFrame. This column should hold the price of the app times the number of installs. What are the top 10 highest grossing paid apps according to this estimate? Out of the top 10 highest grossing paid apps, how many are games?


```python
# convert the price column to numeric: data format is $4.99
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace(",", "").str.replace("$", "")
df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
print(df_apps_clean.Price.describe())
df_apps_clean.sort_values(by="Price", ascending=False).head(10)
```

### The most expensive apps sub $250

```python
# Remove all apps that cost more than $250 from the df_apps_clean DataFrame.
df_apps_clean = df_apps_clean[df_apps_clean.Price <= 250]
print(df_apps_clean.describe())
print(df_apps_clean.head())
df_apps_clean.sort_values(by="Price", ascending=False)
```

### Highest Grossing Paid Apps (ballpark estimate)

```python
# Add a column called 'Revenue_Estimate' to the DataFrame. 
# This column should hold the price of the app times the number of installs.
df_apps_clean["Revenue_Estimate"] = df_apps_clean.Price * df_apps_clean.Installs
# df_apps_clean['Revenue_Estimate'] = df_apps_clean.Installs.mul(df_apps_clean.Price)
df_apps_clean.head()

```

```python
# What are the top 10 highest-grossing paid apps according to this estimate? 
top_grossing = df_apps_clean.sort_values(by="Revenue_Estimate", ascending=False)[:10]
top_grossing
```

```python
# Out of the top 10, how many are games?
print(top_grossing[top_grossing.Category == "GAME"].shape)
top_grossing[top_grossing.Category == "GAME"]
```

```python
# What matters is not just the total number of apps in the category but how often apps are downloaded in that category.
# This will give us an idea of how popular a category is. 
# First, we have to group all our apps by category and sum the number of installations.
category_installs = df_apps_clean.groupby(by="Category").agg({"Installs": pd.Series.sum})
category_installs.sort_values(by="Installs", ascending=True, inplace=True)
category_installs
```

# Extracting Nested Data from a Column

**Challenge**: How many different types of genres are there? Can an app belong to more than one genre? Check what happens when you use .value_counts() on a column with nested values? See if you can work around this problem by using the .split() function and the DataFrame's [.stack() method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.stack.html). 


```python
# number of genres
df_apps_clean.Genres.value_counts() # problem: have number of genres separated by ;
```

```python
# After we’ve separated our genre names based on the semi-colon, 
# we can add them all into a single column with .stack() and then use .value_counts()

# Split the strings on the semi-colon and then .stack them.
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()
print(f'We now have a single column with shape: {stack.shape}')
```

```python
print(stack.index)
print(stack.values)
```

```python
stack
```

```python
num_genres = stack.value_counts()
print(f'Number of genres: {len(num_genres)}')
num_genres
```
