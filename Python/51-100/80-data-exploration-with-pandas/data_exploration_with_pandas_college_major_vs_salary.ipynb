# Cleaning Data

```python
# looking for missing/junk data
# NaN(Not A Number) are blank cell/cells that contain strings instead of numbers.
df.isna()
```

# Sorting Values & Adding Column

```python
# Lowest Risk Majors: lowest difference between 10th and 90th percentile salary

# difference between two columns
clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
# alternate method: clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])

# output is another pandas dataframe column
```

```python
# Add the spread/difference column to the clean_df
spread_col = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]

# insert column into clean_df
clean_df.insert(loc=1, column="Spread", value=spread_col)
clean_df.head()
```

```python
# sorting by lowest spread
low_risk = clean_df.sort_values(by="Spread", ascending=True)

# display the two columns only
low_risk[["Undergraduate Major", "Spread"]].head()
```

```python
# get the top 5 degrees with the highest values in the 90th percentile.

high_potential =  clean_df.sort_values(by="Mid-Career 90th Percentile Salary", ascending=False)
high_potential[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

```python
# get the degrees with the greatest spread in salaries
high_spread = clean_df.sort_values(by="Spread", ascending=False)
high_spread[["Undergraduate Major", "Spread"]].head()
```

```python
# get the degrees where median salaries are high
high_median = clean_df.sort_values(by="Mid-Career Median Salary", ascending=False)
high_median[["Undergraduate Major", "Mid-Career Median Salary"]].head()
```

# Grouping and pivoting data with pandas

```python
# to find the total rows that belongs to a particular group
# Here there are three groups: Business, HASS(Humanities, Arts and Social Science), STEM(Science, Technology, Engineering, Mathematics)
clean_df.groupby("Group").count()
```

```python
# mean of groups
clean_df.groupby("Group").mean()
```

```python
# average salary by group
clean_df.groupby("Group").mean("Mid-Career Median Salary").sort_values(by="Mid-Career Median Salary")
```

```python
# change the number format
pd.options.display.float_format = "{:,.2f}".format
clean_df.groupby("Group").mean()
```
